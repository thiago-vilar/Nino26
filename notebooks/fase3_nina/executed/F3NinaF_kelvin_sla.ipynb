## 1. Contexto e delimitação científica

## 2. Pergunta científica

## 3. Hipótese específica

A sequência de vento, SLA/SSH e termoclina durante La Niña apresenta polaridade e ordenamento temporal coerentes com descarga/upwelling e soerguimento da termoclina, distinguindo gênese, crescimento, pico e decaimento.

A hipótese poderá ser sustentada, parcialmente sustentada ou rejeitada; a execução do notebook não antecipa o resultado.
