# Arquivo legado desativado - 3B_alvo_eventos_ciclo_vida

Este notebook fica em `_broken` apenas como marcador historico local. Ele nao deve ser executado nem citado como saida da Fase 3.

Use a versao ativa em `../3B_alvo_eventos_ciclo_vida.ipynb`.

Contrato metodologico atual da Fase 3:

- evento El Nino: criterio NOAA/ONI local aplicado a OISST do projeto, com media movel de 3 meses da SSTA Nino 3.4 >= +0,5 C por 5 estacoes moveis sobrepostas;
- intensidade: `fraco`, `moderado`, `forte` e `muito_forte`, pelo pico ONI local;
- DHW canonico: `dhw_cweek_0p5_12w`, HotSpots diarios SSTA >= +0,5 C acumulados em 12 semanas;
- vento/acoplamento local: `tau_x_anom_nino34_pa`;
- eixos longitudinais: referencias oficiais W/E, com leitura oeste -> leste.
