## Descritivo (por que este notebook existe)
Testa se RF/XGBoost com XAI caracterizam o ciclo melhor que a Fase 3 e que os baselines — só se justifica se houver ganho real fora da amostra.

## Pergunta
O ML acrescenta capacidade fora da amostra, além da estatística, da climatologia e da persistência, na caracterização das fases?

## Desafio (hipótese a testar)
A persistência semana-a-semana é fortíssima; validação por evento e preprocessing dentro do fold são obrigatórios, senão o F1 é ilusório.

## Metodologia (com referências)
Janela deslizante (lags 4-52), classificação das 4 fases com folds por evento, RFECV aninhado, SHAP/PDP; baselines persistência e semana-do-ano (Cawley & Talbot, 2010; Roberts et al., 2017; Ham et al., 2019; Bracco et al., 2024).
