## Descritivo (por que este notebook existe)
Mede quais variáveis do Pacífico determinam cada fase, por efeito não-paramétrico e por regressão linear multivariada — a leitura estrutural antes do sinal espacial.

## Pergunta
Quais variáveis do Pacífico mais determinam gênese, crescimento, pico e decaimento?

## Desafio (hipótese a testar)
Coeficientes precisam de incerteza por evento (HAC/Newey-West); barras cruzando zero não são distinguíveis de nulo.

## Metodologia (com referências)
Friedman pareado por evento/Kendall W e MLR padronizada com incerteza por evento; controle de multiplicidade e significância de campo (Wilks, 2016; Bretherton et al., 1999).
